```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data
data = {
    "Engine_Size": [1.6, 2.0, 2.5, 3.0, 1.8],
    "Price": [20000, 25000, 30000, 35000, 22000]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Define feature names
feature1 = "Engine_Size"
feature2 = "Price"

# Calculate correlation
correlation_value = df[feature1].corr(df[feature2])

# Print the correlation
print(f"Correlation between {feature1} and {feature2}: {correlation_value:.2f}")

# Calculate the correlation matrix
corr = df.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```
